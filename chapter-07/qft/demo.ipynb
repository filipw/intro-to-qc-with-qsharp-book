# Quantum Fourier Transform

The QFT maps computational basis states to phase-encoded states. This notebook demonstrates both manual and framework-based implementations in big-endian and little-endian conventions, using `DumpMachine()` to inspect the resulting quantum states.

```python
import qsharp
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import DumpFourQubitQFT_BigEndian_Manual, DumpQFT_BigEndian_Manual
from qsharp.code import DumpQFT_BigEndian_Framework
from qsharp.code import DumpFourQubitQFT_LittleEndian_Manual, DumpQFT_LittleEndian_Framework
```

## Big-Endian QFT

### 4-qubit manual implementation (value = 10)

```python
DumpFourQubitQFT_BigEndian_Manual(10)
```

### Generalized manual implementation (value = 10)

```python
DumpQFT_BigEndian_Manual(10, 4)
```

### Framework implementation using ApplyQFT (value = 10)

```python
DumpQFT_BigEndian_Framework(10, 4)
```

## Little-Endian QFT

### 4-qubit manual implementation (value = 10)

```python
DumpFourQubitQFT_LittleEndian_Manual(10)
```

### Framework implementation using ApplyQFT (value = 10)

```python
DumpQFT_LittleEndian_Framework(10, 4)
```

## Comparing across different input values

Apply the generalized big-endian manual QFT to several input values to see how the output phases change.

```python
for value in [0, 1, 5, 10, 15]:
    print(f"--- QFT of |{value}> (4 qubits, big-endian) ---")
    DumpQFT_BigEndian_Manual(value, 4)
    print()
```
