# Grover's Search Algorithm

Grover's algorithm finds a marked item in an unstructured search space of $N = 2^n$ items using $O(\sqrt{N})$ oracle queries, compared to $O(N)$ classically.

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import RunGrover
```

```python
bit_size = 8
targets = [5, 42, 100, 200]
shots = 100

fig, axes = plt.subplots(1, len(targets), figsize=(16, 4))

for ax, target in zip(axes, targets):
    results = [RunGrover(target, bit_size) for _ in range(shots)]
    counts = {}
    for v in results:
        counts[v] = counts.get(v, 0) + 1

    sorted_items = sorted(counts.items())
    labels = [str(k) for k, _ in sorted_items]
    values = [v for _, v in sorted_items]
    colors = ["green" if int(l) == target else "steelblue" for l in labels]
    ax.bar(labels, values, color=colors)
    ax.set_title(f"target = {target}")
    ax.set_ylabel("Counts")
    ax.set_ylim(0, shots + 5)

plt.suptitle(f"Grover's Search ({bit_size}-bit, {shots} shots)")
plt.tight_layout()
plt.show()
```
