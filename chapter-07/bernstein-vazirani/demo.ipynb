# Bernstein-Vazirani Algorithm

The Bernstein-Vazirani algorithm finds a hidden secret string $s$ encoded in a function $f(x) = s \cdot x \pmod{2}$ using a single query to the oracle, compared to $n$ queries classically.

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import RunBernsteinVazirani
```

```python
bit_size = 3
shots = 100

fig, axes = plt.subplots(2, 4, figsize=(16, 7))

for secret in range(2**bit_size):
    ax = axes[secret // 4, secret % 4]
    results = []
    for _ in range(shots):
        r = RunBernsteinVazirani(secret, bit_size)
        bits = [1 if str(b) == "One" else 0 for b in r]
        value = sum(b * 2**i for i, b in enumerate(bits))
        results.append(value)

    counts = {}
    for v in results:
        counts[v] = counts.get(v, 0) + 1

    labels = [str(k) for k in sorted(counts.keys())]
    values = [counts[int(k)] for k in labels]
    ax.bar(labels, values)
    ax.set_title(f"secret = {secret}")
    ax.set_ylabel("Counts")
    ax.set_ylim(0, shots + 10)

plt.suptitle("Bernstein-Vazirani: recovered secret values")
plt.tight_layout()
plt.show()
```
