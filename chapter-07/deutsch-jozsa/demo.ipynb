# Deutsch-Jozsa Algorithm

The Deutsch-Jozsa algorithm determines whether a function is **constant** (same output for all inputs) or **balanced** (outputs 0 for half the inputs and 1 for the other half) using a single query.

- All-zero measurement result: **constant**
- Any non-zero measurement result: **balanced**

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import RunDeutschJozsa
```

```python
oracle_names = {
    0: "Constant Zero",
    1: "Constant One",
    2: "Balanced Equal",
    3: "Balanced Not-Equal",
}

shots = 100
n_values = [1, 3, 6]

fig, axes = plt.subplots(len(n_values), 4, figsize=(16, 3 * len(n_values)))

for row, n in enumerate(n_values):
    for col, oracle_type in enumerate(range(4)):
        ax = axes[row, col]
        classifications = {"Constant": 0, "Balanced": 0}
        for _ in range(shots):
            is_constant = RunDeutschJozsa(n, oracle_type)
            if is_constant:
                classifications["Constant"] += 1
            else:
                classifications["Balanced"] += 1

        ax.bar(classifications.keys(), classifications.values())
        ax.set_title(f"n={n}, {oracle_names[oracle_type]}")
        ax.set_ylabel("Counts")
        ax.set_ylim(0, shots + 10)

plt.suptitle("Deutsch-Jozsa: classification results")
plt.tight_layout()
plt.show()
```
