# Basic Qubits

This notebook demonstrates qubit allocation and measurement in different bases.

- A qubit in the default state $|0\rangle$ measured in the **Z basis** always yields `Zero`.
- The same qubit measured in the **X basis** yields `Zero` or `One` with equal probability.
- A qubit in **superposition** ($H|0\rangle = |+\rangle$) measured in the **Z basis** yields `Zero` or `One` with equal probability.
- The same superposition qubit measured in the **X basis** always yields `Zero` (since $|+\rangle$ is an eigenstate of X).

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import MeasureDefaultQubitZ, MeasureDefaultQubitX
from qsharp.code import MeasureSuperpositionQubitZ, MeasureSuperpositionQubitX
```

```python
shots = 4096

experiments = {
    "Default qubit, Z basis": MeasureDefaultQubitZ,
    "Default qubit, X basis": MeasureDefaultQubitX,
    "Superposition qubit, Z basis": MeasureSuperpositionQubitZ,
    "Superposition qubit, X basis": MeasureSuperpositionQubitX,
}

fig, axes = plt.subplots(1, 4, figsize=(16, 4))

for ax, (title, op) in zip(axes, experiments.items()):
    results = [str(op()) for _ in range(shots)]
    counts = {"Zero": results.count("Zero"), "One": results.count("One")}
    ax.bar(counts.keys(), counts.values())
    ax.set_title(title)
    ax.set_ylabel("Counts")
    ax.set_ylim(0, shots)

plt.tight_layout()
plt.show()
```
