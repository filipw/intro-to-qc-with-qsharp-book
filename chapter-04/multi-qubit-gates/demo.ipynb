# Multi-Qubit Gates

This notebook demonstrates CNOT, SWAP, and CCNOT (Toffoli) gates by building truth tables.
Each gate is shown in both its library and manual (decomposed) form.

```python
import qsharp
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import LibraryCNOT, ManualCNOT
from qsharp.code import LibrarySWAP, ManualSWAP
from qsharp.code import LibraryCCNOT, ManualCCNOT
```

```python
def result_to_str(results):
    return "".join("1" if str(r) == "One" else "0" for r in results)

def truth_table(name, op, n_qubits):
    print(f"  {name}")
    print(f"  {'Input':>8}  |  Output")
    print(f"  {'-'*8}--+--{'-'*8}")
    for i in range(2**n_qubits):
        bits = [bool((i >> (n_qubits - 1 - j)) & 1) for j in range(n_qubits)]
        result = op(bits)
        inp = "".join("1" if b else "0" for b in bits)
        out = result_to_str(result)
        print(f"  |{inp}>     |  |{out}>")
    print()
```

## CNOT

Flips the target qubit if the control qubit is $|1\rangle$.

```python
truth_table("Library CNOT", LibraryCNOT, 2)
truth_table("Manual CNOT (Controlled X)", ManualCNOT, 2)
```

## SWAP

Exchanges the states of two qubits.

```python
truth_table("Library SWAP", LibrarySWAP, 2)
truth_table("Manual SWAP (3 CNOTs)", ManualSWAP, 2)
```

## CCNOT (Toffoli)

Flips the target qubit only when both control qubits are $|1\rangle$.

```python
truth_table("Library CCNOT", LibraryCCNOT, 3)
truth_table("Manual CCNOT (Controlled Controlled X)", ManualCCNOT, 3)
```
