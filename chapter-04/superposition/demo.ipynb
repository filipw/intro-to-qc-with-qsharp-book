# Superposition and Random Number Generation

This notebook demonstrates three approaches to quantum random number generation using superposition:

1. **Single-qubit method** - measures one qubit at a time, builds the number bit by bit.
2. **N-qubit method** - puts N qubits in superposition simultaneously, measures all at once.
3. **Framework method** - uses the built-in `MeasureInteger` operation.

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import RandomNumber1Qubit, RandomNumberNQubits, RandomNumberFramework
```

```python
shots = 4096
bit_count = 8

methods = {
    "Single qubit": RandomNumber1Qubit,
    "N qubits": RandomNumberNQubits,
    "Framework": RandomNumberFramework,
}

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

for ax, (title, op) in zip(axes, methods.items()):
    results = [op(bit_count) for _ in range(shots)]
    ax.hist(results, bins=range(2**bit_count + 1), edgecolor="black", linewidth=0.5)
    ax.set_title(title)
    ax.set_xlabel("Value")
    ax.set_ylabel("Counts")

plt.tight_layout()
plt.show()
```
