# Superdense Coding

Superdense coding allows Alice to send **2 classical bits** by transmitting only **1 qubit**, using a pre-shared entangled pair.

We verify that for all 4 possible inputs, Bob correctly decodes Alice's message with 100% accuracy.

```python
import qsharp
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import RunDenseCoding
```

```python
shots = 1024

for v1 in [False, True]:
    for v2 in [False, True]:
        successes = 0
        for _ in range(shots):
            b1, b2 = RunDenseCoding(v1, v2)
            if b1 == v1 and b2 == v2:
                successes += 1
        bit1 = "1" if v1 else "0"
        bit2 = "1" if v2 else "0"
        print(f"Input ({bit1},{bit2}): success rate = {successes * 100 / shots:.1f}%")
```
