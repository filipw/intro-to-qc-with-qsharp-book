# Entanglement and Bell States

This notebook creates Bell states from different initial states and measures in different bases.

When two qubits are entangled and measured in the **same basis** (Z,Z), the outcomes are perfectly correlated.
When measured in **different bases** (Z,X), the correlation disappears.

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import BellState
```

```python
shots = 4096
labels = ["|00>", "|01>", "|10>", "|11>"]

experiments = [
    ("Init |00>, ZZ basis", False, False, True, True),
    ("Init |01>, ZZ basis", False, True, True, True),
    ("Init |10>, ZZ basis", True, False, True, True),
    ("Init |11>, ZZ basis", True, True, True, True),
    ("Init |00>, ZX basis", False, False, True, False),
]

fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for ax, (title, i0, i1, b0, b1) in zip(axes, experiments):
    counts = [0, 0, 0, 0]
    for _ in range(shots):
        r = BellState(i0, i1, b0, b1)
        idx = (1 if str(r[0]) == "One" else 0) * 2 + (1 if str(r[1]) == "One" else 0)
        counts[idx] += 1
    ax.bar(labels, counts)
    ax.set_title(title)
    ax.set_ylabel("Counts")
    ax.set_ylim(0, shots)

plt.tight_layout()
plt.show()
```
