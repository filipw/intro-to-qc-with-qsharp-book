# GHZ Game

The GHZ game is a three-player nonlocal game. The winning condition is:

$a \oplus b \oplus c = x \lor y \lor z$

The valid inputs satisfy $x \oplus y \oplus z = 0$.

- The best **classical** strategy wins at most **75%** (3 out of 4 input combinations).
- A **quantum** strategy using a GHZ state wins **100%** of the time.

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import QuantumGHZStrategy, ClassicalStrategy, CheckGameOutcome
```

```python
inputs = [
    [False, False, False],
    [False, True, True],
    [True, False, True],
    [True, True, False],
]

rounds_per_input = 128
q_wins = 0
c_wins = 0
total = len(inputs) * rounds_per_input

for xyz in inputs:
    q_round = 0
    c_round = 0
    for _ in range(rounds_per_input):
        q_out = QuantumGHZStrategy(xyz[0], xyz[1], xyz[2])
        if CheckGameOutcome(xyz, q_out):
            q_round += 1

        c_out = ClassicalStrategy(xyz[0], xyz[1], xyz[2])
        if CheckGameOutcome(xyz, c_out):
            c_round += 1

    q_wins += q_round
    c_wins += c_round
    print(f"Input {xyz}: Quantum won {q_round}/{rounds_per_input}, Classical won {c_round}/{rounds_per_input}")

print(f"\nOverall: Quantum {q_wins * 100 / total:.1f}%, Classical {c_wins * 100 / total:.1f}%")

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(["Classical", "Quantum"], [c_wins * 100 / total, q_wins * 100 / total])
ax.axhline(y=75, color="r", linestyle="--", label="Classical bound (75%)")
ax.set_ylabel("Win rate (%)")
ax.set_ylim(0, 110)
ax.legend()
plt.title("GHZ Game")
plt.show()
```
