# Entanglement Swapping

Entanglement swapping allows two qubits (q1 and q3) that **never interacted** to become entangled, by performing a Bell measurement on their respective partners (q2 and q4).

After the protocol, q1 and q3 are in the Bell state $|\Phi^+\rangle$, so their measurements are perfectly correlated: we expect ~50% $|00\rangle$ and ~50% $|11\rangle$.

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import EntanglementSwapping
```

```python
shots = 4096
labels = ["|00>", "|01>", "|10>", "|11>"]
counts = [0, 0, 0, 0]

for _ in range(shots):
    r = EntanglementSwapping()
    r0 = 1 if str(r[0]) == "One" else 0
    r1 = 1 if str(r[1]) == "One" else 0
    counts[r0 * 2 + r1] += 1

matching = counts[0] + counts[3]
print(f"Correlated outcomes (|00> + |11>): {matching}/{shots} ({matching*100/shots:.1f}%)")

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(labels, counts)
ax.set_ylabel("Counts")
ax.set_ylim(0, shots)
plt.title("Entanglement Swapping - q1 and q3 outcomes")
plt.show()
```
