# CHSH Game

The CHSH game demonstrates quantum advantage: a classical strategy can win at most **75%** of the time, while a quantum strategy using entanglement can achieve approximately **cos²(π/8) ≈ 85%**.

The winning condition is: $(a \cdot b) = (x \oplus y)$ where $a,b$ are the referee's bits and $x,y$ are the players' outputs.

```python
import qsharp
import matplotlib.pyplot as plt
import random
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import ClassicalStrategy, QuantumStrategy
```

```python
shots = 4096

def run_game(strategy, n):
    wins = 0
    for _ in range(n):
        a = random.choice([True, False])
        b = random.choice([True, False])
        x, y = strategy(a, b)
        if (a and b) == (x ^ y):
            wins += 1
    return wins / n

classical_rate = run_game(ClassicalStrategy, shots)
quantum_rate = run_game(QuantumStrategy, shots)

print(f"Classical win rate: {classical_rate * 100:.2f}%")
print(f"Quantum win rate:   {quantum_rate * 100:.2f}%")

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(["Classical", "Quantum"], [classical_rate * 100, quantum_rate * 100])
ax.axhline(y=75, color="r", linestyle="--", label="Classical bound (75%)")
ax.set_ylabel("Win rate (%)")
ax.set_ylim(0, 100)
ax.legend()
plt.title("CHSH Game")
plt.show()
```
