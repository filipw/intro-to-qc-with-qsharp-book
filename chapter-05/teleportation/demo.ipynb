# Quantum Teleportation

Teleportation transfers an arbitrary quantum state from one qubit to another using an entangled pair and two classical bits of communication.

We verify correctness by preparing a known state, teleporting it, then applying the inverse preparation and checking that we get $|0\rangle$ (measurement result `Zero`).

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import RunTeleportation
```

```python
shots = 4096
results = [str(RunTeleportation()) for _ in range(shots)]
successes = results.count("Zero")

print(f"Success rate: {successes * 100 / shots:.2f}% ({successes}/{shots})")

fig, ax = plt.subplots(figsize=(5, 4))
ax.bar(["Success (Zero)", "Failure (One)"], [successes, shots - successes])
ax.set_ylabel("Counts")
ax.set_ylim(0, shots * 1.1)
plt.title("Teleportation Verification")
plt.show()
```
