# Bell's Inequality and CHSH

## Bell's inequality
For three measurement settings (a, b, c), classical correlations satisfy:
$P(a+,b+) + P(b+,c+) \geq P(a+,c+)$

Quantum mechanics violates this.

## CHSH inequality
$|E(a_1,b_1) + E(a_1,b_2)| + |E(a_2,b_1) - E(a_2,b_2)| \leq 2$

Quantum mechanics reaches up to $2\sqrt{2} \approx 2.83$.

```python
import qsharp
import matplotlib.pyplot as plt
```

```python
qsharp.init()
with open("Program.qs") as f:
    qsharp.eval(f.read())

from qsharp.code import BellAB, BellBC, BellAC
from qsharp.code import CHSH_A1B1, CHSH_A1B2, CHSH_A2B1, CHSH_A2B2
```

## Bell's Inequality Test

```python
shots = 4096

def prob_both_zero(op, n):
    count = sum(1 for _ in range(n) if str(op()[0]) == "Zero" and str(op()[1]) == "Zero")
    return count / n

pAB = prob_both_zero(BellAB, shots)
pBC = prob_both_zero(BellBC, shots)
pAC = prob_both_zero(BellAC, shots)

print(f"P(a+,b+) = {pAB:.4f}")
print(f"P(b+,c+) = {pBC:.4f}")
print(f"P(a+,c+) = {pAC:.4f}")
print(f"\nBell inequality: P(AB) + P(BC) >= P(AC)")
print(f"{pAB:.4f} + {pBC:.4f} = {pAB + pBC:.4f} >= {pAC:.4f} ? {pAB + pBC >= pAC}")
```

## CHSH Inequality Test

```python
def correlator(op, n):
    counts = [0, 0, 0, 0]
    for _ in range(n):
        r = op()
        r0 = 1 if str(r[0]) == "One" else 0
        r1 = 1 if str(r[1]) == "One" else 0
        counts[r0 * 2 + r1] += 1
    return (counts[0] + counts[3] - counts[1] - counts[2]) / n

e11 = correlator(CHSH_A1B1, shots)
e12 = correlator(CHSH_A1B2, shots)
e21 = correlator(CHSH_A2B1, shots)
e22 = correlator(CHSH_A2B2, shots)

S = abs(e11 + e12) + abs(e21 - e22)

print(f"E(a1,b1) = {e11:.4f}")
print(f"E(a1,b2) = {e12:.4f}")
print(f"E(a2,b1) = {e21:.4f}")
print(f"E(a2,b2) = {e22:.4f}")
print(f"\nCHSH value S = {S:.4f}")
print(f"Classical bound: S <= 2.0")
print(f"CHSH inequality satisfied? {S <= 2.0}")
```
